## 如何交互式地运行lammps

为了更加方便地在`jupyter`中运行lammps, 我们对lammps的python api做了绑定和注册. 现在, 你可以直接在notebook中完成lammps的脚本的编写和一键运行.

首先, 需要编译lammps的python版本, 具体可见: [编译与安装](/docs/Tutorial/install.md)
其次, `pip install -e .` 安装本项目的python包
最后, 在notebook中运行如下代码, 即可完成lammps的交互式运行

```python
# 加载模块
%load_ext lammpscn.lammps
```

```python
# 如果修改了代码，需要重新加载模块
%reload_ext lammpscn.lammps
```

```python
# 执行单行命令, 单个%开头且不换行
%cmd variable a equal 1
```

执行多行命令, 以%%开头且换行

```python
%%cmd
variable a equal 1
variable b equal 2
```
